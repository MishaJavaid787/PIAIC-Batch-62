```python
import google.generativeai as genai
```

Setting Up API Key

```python
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

Using Image as Prompt

```python
!curl -o image.jpg "https://www.apriorit.com/wp-content/uploads/2021/05/AI_for_automotive-01.jpg"
```

```python
import PIL.Image
image = PIL.Image.open('image.jpg')
image
```

```python
prompt = """This image contains a figure describing potential use of AI in Automative Industry.
Given the figure, describe these AI applications as thoroughly as possible based on what you
see in the image, making sure to note all. Return output in json format:
{description: description, applications: [application1, application2, application3, etc]}"""
```

Setting parameters: generating text through text-image input

```python
model = genai.GenerativeModel('gemini-1.5-flash')
response = model.generate_content([prompt, image])
print(response.text)
```
